```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/imdb-top-250-movies/movies.csv')

df.head(2)
```

### 1. Sorting Values based on Duration

```python
duration = []
for i in df['duration']:
    if ('Nan' != i):
        duration.append(int(i))
    else:
        duration.append(100)
    
df['duration'] = duration

df.sort_values(by = 'duration', ascending = False).head()
```

### 2. Sorting the movies based on Year

```python
df.sort_values(by = 'year', ascending = False)
```

### 3. Finding the Unique Genre

```python
gen = []
for i in df['genre']:
    gen += i.split(',')
    
genre = list(set(gen))
```

### 4. Finding the Movie Distribution in each Genre

```python
data = []

for gen in genre:
    c = 0
    for i in df['genre']:
        if (gen in i):
            c += 1
    data.append([gen, c])
print(data)
```

### 5. Sort the Data

```python
df_ = pd.DataFrame(data, columns = ['genre','freq'])
df_.sort_values('freq', ascending = False)
```

```python
df_ = pd.DataFrame(data, columns = ['genre','freq'])

df_.sort_values('freq', ascending = False)
df_.to_csv('genre_dist.csv', index = False)
```
