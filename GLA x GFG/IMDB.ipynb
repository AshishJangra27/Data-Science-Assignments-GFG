```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

link = 'https://www.imdb.com/chart/top/'

res = requests.get(link)
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
data = []

for sp in soup.find('tbody', class_ = 'lister-list').find_all('tr'):

  movie_link =  'https://www.imdb.com' +  sp.find('td', class_ = 'posterColumn').find('a').get('href')
  poster_link = sp.find('td', class_ = 'posterColumn').find('img').get('src')
  cast = sp.find('td', class_ = 'titleColumn').find('a').get('title')
  year = sp.find('td', class_ = 'titleColumn').find('span').text
  
  # print(movie_link)
  # print(poster_link)
  # print(cast)
  # print(year)

  # print('--')

  data.append([year,cast,movie_link,poster_link])

  # break
```

```python
df = pd.DataFrame(data, columns = ['year','cast','movie_link','poster_link'])
df
```

```python
for link in df['movie_link']:

  res = requests.get(link)
  sp = BeautifulSoup(res.content, 'html.parser')

  print(sp.find('h1'))

  break
```

```python
df.to_csv('movies.csv', index = False)
```
